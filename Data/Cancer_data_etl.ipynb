```python
import json
import pandas as pd
```

```python
#import incidence csv file
incidence_df = pd.read_csv("incidence.csv")
```

```python
incidence_df.head()
```

```python
incidence_df.columns
```

```python
#keep only the count columns that we want
incidence_revised_df =incidence_df[['State_code', 'State', ' FIPS', 'bladder_count',
       'brain_count', 'breast_count', 'cervix_count', 'colon_rectum_count',
       'esophagus_count', 'kidney_count', 'leukemia_count', 'liver_count',
       'lung_count', 'non_HodgkinL_count', 'oral_pharynx_count', 'ovary_count',
       'pancreas_count', 'prostate_count', 'melanoma_count', 'stomach_count',
        'thyroid_count', 'uterus_count']]
```

```python
incidence_revised_df
```

```python
#remove Washington DC and Puerto Rico
incidence_revised_df = incidence_revised_df.drop([9,52])
incidence_revised_df
```

```python
#create total incidence column
col_list = list(incidence_revised_df)
col_list.remove(' FIPS') 
#col_list
incidence_revised_df["all_cancers"] = incidence_revised_df[col_list].sum(axis=1)
incidence_revised_df
```

```python
#reorder columns
incidence_revised_df = incidence_revised_df[['State_code', 'State', 'all_cancers', ' FIPS', 'bladder_count',
       'brain_count', 'breast_count', 'cervix_count', 'colon_rectum_count',
       'esophagus_count', 'kidney_count', 'leukemia_count', 'liver_count',
       'lung_count', 'non_HodgkinL_count', 'oral_pharynx_count', 'ovary_count',
       'pancreas_count', 'prostate_count', 'melanoma_count', 'stomach_count',
        'thyroid_count', 'uterus_count']]
```

```python
#convert to json file
incidence_revised_df.to_json(r'cancer_incidence.json')
```

```python
#convert to json file
incidence_revised_df.to_json(r'cancer_incidence.json', orient = "index")
```
